TEXT TRANSLATION

```python
from transformers import pipeline

# Define a dictionary of supported translation pipelines by target language code
translation_models = {
    'fr': 'Helsinki-NLP/opus-mt-en-fr',  # English to French
    'es': 'Helsinki-NLP/opus-mt-en-es',  # English to Spanish
    'de': 'Helsinki-NLP/opus-mt-en-de',  # English to German
    'it': 'Helsinki-NLP/opus-mt-en-it',  # English to Italian
    # Add more if needed
}

# Ask user for target language code
target_lang = input("Enter target language code (e.g., fr, es, de, it): ").strip().lower()

if target_lang in translation_models:
    translator = pipeline("translation_en_to_" + target_lang, model=translation_models[target_lang])

    translation = translator(text, max_length=512)
    translated_text = translation[0]['translation_text']

    print(f"Translated Text ({target_lang}):")
    print(translated_text)
else:
    print("Sorry, that language is not supported yet.")
```

SUMMARIZE

```python
from transformers import pipeline

# Load summarization pipeline
textsummarize = pipeline('summarization')

# Summarize the transcribed text
summary = textsummarize(text, max_length=100, min_length=25, do_sample=False)

print("Summary:")
print(summary[0]['summary_text'])
```

QUESTION-ANSWER

```python
from transformers import pipeline

# Use a stronger Q&A model
q_a = pipeline("question-answering", model="deepset/roberta-base-squad2")

# Use the summarized text as context (replace this with your actual summary variable)
context = summary[0]['summary_text']

print("Context for Q&A:", context)

while True:
    question = input("\nEnter your question (or type 'exit' to quit): ")
    if question.lower() == 'exit':
        print("Exiting the Q&A session.")
        break

    # Make sure question and context are not empty or too short
    if not question.strip():
        print("Please enter a valid question.")
        continue

    result = q_a(question=question, context=context)
    print("Answer:", result['answer'])
```

SENTIMENT ANALYSIS

```python
from transformers import pipeline

# Load sentiment-analysis pipeline (default model: distilbert-base-uncased-finetuned-sst-2-english)
sentiment_analyzer = pipeline("sentiment-analysis")

# Example: analyze the sentiment of the same context or any text
text_to_analyze = context  # you can also use `text` or `summary[0]['summary_text']`

result = sentiment_analyzer(text_to_analyze)

print("Sentiment Analysis Result:")
print(f"Label: {result[0]['label']}, Score: {result[0]['score']:.4f}")
```
